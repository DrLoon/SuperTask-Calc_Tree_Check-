**7.** Дана функция:

$$
 \begin{cases}
    x^2 + 2x + 6, & x < 0  \\
    x + 6, & 0 \le x \le 2 \\
    x^2 + 4x - 4, & x \ge 2
 \end{cases}
$$

Постройте массив из её значений на  $-3 \le x \le 3$.

```python
def f7(x):
    if x < 0:
        return x*x + 2*x + 6
    elif x <= 2:
        return x + 6
    else:
        return x*x + 4*x - 4
X7 = np.linspace(-3,3,100)
fvec7 = np.vectorize(f7)
Y7 = fvec7(X7)
Y7
```

**10.** Создать матрицу:

$$
\begin{pmatrix}
0 & 1 & 2 & 3 & 4 \\
1 & 2 & 3 & 4 & 0 \\
2 & 3 & 4 & 0 & 1 \\
3 & 4 & 0 & 1 & 2 \\
4 & 0 & 1 & 2 & 3
\end{pmatrix}
$$

```python
n = 5
r = np.arange(0,n)
X10 = np.tile(r, (n,1))
rows, column_indices = np.ogrid[:n, :n]
r *= -1
r[r < 0] += n
column_indices = column_indices - r[:,np.newaxis]
result = X10[rows, column_indices]
print(result)
```

**11.** Есть следующий алгоритм семплинирования, на вход функции подается вектор из целых положительных чисел и число семплов.

```python
def sample(x, c):
    assert len(x) > 0
    
    s = np.sum(x)
    res = []
    for i in range(c):
        val = s * np.random.random()
        cur, idx = 0, 0        
        while cur + x[idx] <= val:
            cur += x[idx]
            idx += 1
        res.append(idx)
    return res

            
print(sample([50, 3, 1, 7, 20], 5))
```

Реализуйте данный алгоритм более оптимальным способом, без использования циклов.

```python
def sample(x, c):
    assert len(x) > 0
    
    s = np.sum(x)

    def f11(ranf):
        val = s * ranf
        cusum = np.cumsum(x)
        idx = np.searchsorted(cusum, val)
        return idx

    agr = np.random.random(c)
    g = f11(agr)
    return list(g)

            
print(sample([50, 3, 1, 7, 20], 5))
```
